```
def f(x, y):
  return np.cos(x) ** 10 + np.sin(10 + y * x) * np.sin(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
plt.contourf(X, Y, f(X, Y), 20, cmap='RdGy')
```
