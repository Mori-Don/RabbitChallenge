```
import pandas as pd
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import seaborn as sns
```

```
plt.style.use('seaborn')

dataset = load_boston()
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
df['MEDV'] = dataset.target
```

```
df
```

```
df.describe()
```

```
sns.heatmap(df.corr())
```

```
sns.pairplot(df[['RM','LSTAT','MEDV']])
```

```
#distplot()関数　・・・　ヒストグラムと密度関数を得られる
sns.distplot(df['MEDV'])
```

```
#jointplot() ・・・ 2つの値の分布について相関を得られる
x, y = df['MEDV'], df['LSTAT']
sns.jointplot(x, y, alpha=0.5)
```

```
#jointplot()の別種の表示方式
sns.jointplot(df['MEDV'], df['LSTAT'], kind='kde')
```

```
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```
iris = datasets.load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['target'] = iris.target_names[iris.target]
data
```

```
x_train, x_test, y_train, y_test = train_test_split(iris['data'], iris['target'])
```

```
data_train = pd.DataFrame(x_train, columns=iris.feature_names)
data_train['target'] = iris.target_names[y_train]
data_train
```

```
data_test = pd.DataFrame(x_test, columns=iris.feature_names)
data_test['target'] = iris.target_names[y_test]
data_test
```

```
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train, y_train)
```

```
predicted = knn.predict([x_test[0]])
predicted
```
